```python
#import packages
import numpy as np
import pandas as pd
import re
import scipy
import datetime
```

```python
#baseline Cre
labdata=pd.read_csv('G:/AKIprediction/labdataaki.csv',encoding='gbk')
CHDdata_AKI=pd.read_csv('G:/AKIprediction/CHDdata_AKI_original.csv',encoding='gbk')
for c in ['ADMITTIME','DISCHARGETIME','surgery_start_time','surgery_end_time']:
    CHDdata_AKI[c]=CHDdata_AKI[c].apply(lambda x:pd.to_datetime(x))
labdata['CHARTTIME']=labdata['CHARTTIME'].apply(lambda x:pd.to_datetime(x))
cre=labdata[labdata.ITEMID==5041]
def baselinecre(hid,starttime,endtime,admittime,dischargetime):
    crep=cre[(cre.HADM_ID==hid)&(cre.CHARTTIME>=admittime)&(cre.CHARTTIME<=dischargetime)&(cre.FLAG=='z')]
    if crep.shape[0]>0:
        crebeforesurgery=crep[(crep.CHARTTIME<=starttime)&(crep.CHARTTIME>=starttime-datetime.timedelta(days=5))]
        creaftersurgery=crep[crep.CHARTTIME>=endtime]
        if crebeforesurgery.shape[0]>0:
            crebeforesurgery.sort_values(by='CHARTTIME',ascending=False,inplace=True)
            crebeforesurgery.reset_index(drop=True,inplace=True)
            return crebeforesurgery.loc[0,'VALUENUM']
        else:
            if creaftersurgery.shape[0]>0:
                creaftersurgery.sort_values(by='CHARTTIME',inplace=True)
                creaftersurgery.reset_index(drop=True,inplace=True)
                return creaftersurgery.loc[0,'VALUENUM']
            else:
                return np.nan
CHDdata_AKI['baselineCre']=list(map(lambda hid,starttime,endtime,admittime,dischargetime:baselinecre(hid,starttime,endtime,admittime,dischargetime),
         CHDdata_AKI['INHOSP_NO'],CHDdata_AKI['surgery_start_time'],CHDdata_AKI['surgery_end_time'],CHDdata_AKI['ADMITTIME'],CHDdata_AKI['DISCHARGETIME']))
```

```python
#any AKI label
cre=labdata[(labdata.ITEMID==5041)|(labdata.ITEMID==5032)|(labdata.ITEMID==6954)]
cre.sort_values(by='CHARTTIME',inplace=True)
def anyaki(sid,basecre,endtime,dischargetime):
    cre1=cre[(cre['SUBJECT_ID']==sid)&(cre['CHARTTIME']<=dischargetime)&(cre['CHARTTIME']>=endtime)]
    if cre1.shape[0]>0:
        cre2=cre1[(cre1['VALUENUM']>=1.5*basecre)]
        if cre2.shape[0]>0:
            cre2.reset_index(drop=True,inplace=True)
            return cre2.loc[0,'CHARTTIME']
        else:
            cre1['VALUE1']=cre1['VALUENUM']-cre1['VALUENUM'].shift(1)
            cre1['time1']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(1)).dt.total_seconds()/3600
            cre1['VALUE2']=cre1['VALUENUM']-cre1['VALUENUM'].shift(2)
            cre1['time2']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(2)).dt.total_seconds()/3600
            cre1['VALUE3']=cre1['VALUENUM']-cre1['VALUENUM'].shift(3)
            cre1['time3']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(3)).dt.total_seconds()/3600
            cre1['VALUE4']=cre1['VALUENUM']-cre1['VALUENUM'].shift(4)
            cre1['time4']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(4)).dt.total_seconds()/3600
            cre1['VALUE5']=cre1['VALUENUM']-cre1['VALUENUM'].shift(5)
            cre1['time5']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(5)).dt.total_seconds()/3600
            cre1['VALUE6']=cre1['VALUENUM']-cre1['VALUENUM'].shift(6)
            cre1['time6']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(6)).dt.total_seconds()/3600
            cre1['VALUE7']=cre1['VALUENUM']-cre1['VALUENUM'].shift(7)
            cre1['time7']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(7)).dt.total_seconds()/3600
            cre1['VALUE8']=cre1['VALUENUM']-cre1['VALUENUM'].shift(8)
            cre1['time8']=(cre1['CHARTTIME']-cre1['CHARTTIME'].shift(8)).dt.total_seconds()/3600
            cre3=cre1[(cre1['VALUE1']>=26.5)&(cre1['time1']<=48)]
            cre4=cre1[(cre1['VALUE2']>=26.5)&(cre1['time2']<=48)]
            cre5=cre1[(cre1['VALUE3']>=26.5)&(cre1['time3']<=48)]
            cre6=cre1[(cre1['VALUE4']>=26.5)&(cre1['time4']<=48)]
            cre7=cre1[(cre1['VALUE5']>=26.5)&(cre1['time5']<=48)]
            cre8=cre1[(cre1['VALUE6']>=26.5)&(cre1['time6']<=48)]
            cre9=cre1[(cre1['VALUE7']>=26.5)&(cre1['time7']<=48)]
            cre10=cre1[(cre1['VALUE8']>=26.5)&(cre1['time8']<=48)]
            if cre3.shape[0]>0:
                cre3.reset_index(drop=True,inplace=True)
                return cre3.loc[0,'CHARTTIME']
            else:
                if cre4.shape[0]>0:
                    cre4.reset_index(drop=True,inplace=True)
                    return cre4.loc[0,'CHARTTIME']
                else:
                    if cre5.shape[0]>0:
                        cre5.reset_index(drop=True,inplace=True)
                        return cre5.loc[0,'CHARTTIME']
                    else:
                        if cre6.shape[0]>0:
                            cre6.reset_index(drop=True,inplace=True)
                            return cre6.loc[0,'CHARTTIME']
                        else:
                            if cre7.shape[0]>0:
                                cre7.reset_index(drop=True,inplace=True)
                                return cre7.loc[0,'CHARTTIME']
                            else:
                                if cre8.shape[0]>0:
                                    cre8.reset_index(drop=True,inplace=True)
                                    return cre8.loc[0,'CHARTTIME']
                                else:
                                    if cre9.shape[0]>0:
                                        cre9.reset_index(drop=True,inplace=True)
                                        return cre9.loc[0,'CHARTTIME']
                                    else:
                                        if cre10.shape[0]>0:
                                            cre10.reset_index(drop=True,inplace=True)
                                            return cre10.loc[0,'CHARTTIME']
                                        else:
                                            return np.nan
            return np.nan
    else:
        return np.nan
CHDdata_AKI['anyAKItime']=list(map(lambda sid,basecre,endtime,dischargetime:anyaki(sid,basecre,endtime,dischargetime),
                               CHDdata_AKI['INHOSP_INDEX_NO'],CHDdata_AKI['baselineCre'],
                            CHDdata_AKI['surgery_end_time'],CHDdata_AKI['DISCHARGETIME']))
```

```python
#random onset time of patients without AKI
for c in ['anyAKItime','surgery_end_time','出院日期']:
    CHDdata_AKI[c]=CHDdata_AKI[c].apply(lambda x:pd.to_datetime(x))
CHDdata_AKI['出院日期']=CHDdata_AKI['出院日期'].apply(lambda x:x.date())
CHDdata_AKI11=CHDdata_AKI[CHDdata_AKI.anyAKItime.isnull()]
CHDdata_AKI22=CHDdata_AKI[~CHDdata_AKI.anyAKItime.isnull()]
def AKItimeinterval(endtime,dischargetime):
    global time
    time=np.random.normal(41.5,20.6)
    time=round(time,1)
    if (endtime+datetime.timedelta(hours=time)).date()<=dischargetime:
        return endtime+datetime.timedelta(hours=time)
    else:
        return pd.to_datetime(dischargetime)
CHDdata_AKI11['onsettime']=list(map(lambda endtime,dischargetime:AKItimeinterval(endtime,dischargetime),
         CHDdata_AKI11['surgery_end_time'],CHDdata_AKI11['出院日期']))
CHDdata_AKI22['onsettime']=CHDdata_AKI22['anyAKItime']
CHDdata_AKI=pd.concat([CHDdata_AKI11,CHDdata_AKI22])
```

```python
#perspective 3 random prediction time
CHDdata_AKI['deltatime']=(CHDdata_AKI['onsettime']-CHDdata_AKI['surgery_end_time']).apply(lambda x:x.total_seconds()/3600)
def randomprediction(dtime,endtime):
    global time
    time=np.random.uniform(0,dtime)
    time=round(time,1)
    return endtime+datetime.timedelta(hours=time)
CHDdata_AKI['randomtime']=list(map(lambda dtime,endtime:randomprediction(dtime,endtime),
         CHDdata_AKI['deltatime'],CHDdata_AKI['surgery_end_time']))
```

```python
#surgery_vital_sign
CHDdata_AKI['手术日期']=CHDdata_AKI['手术日期'].apply(lambda x:pd.to_datetime(x).date())
surgery_vital_sign.drop_duplicates(subset=['SUBJECT_ID','MONITOR_TIME','ITEMID'],inplace=True)
surgery_vital_sign['MONITOR_TIME']=surgery_vital_sign['MONITOR_TIME'].apply(lambda x:pd.to_datetime(x))
surgery_vital_sign['MONITOR_DATE']=surgery_vital_sign['MONITOR_TIME'].apply(lambda x:x.date())
surgery_vital_sign.sort_values(by='MONITOR_TIME',inplace=True)
sbp=surgery_vital_sign[surgery_vital_sign.ITEMID==65]
dbp=surgery_vital_sign[surgery_vital_sign.ITEMID==66]
cvp=surgery_vital_sign[surgery_vital_sign.ITEMID==71]
temperature=surgery_vital_sign[surgery_vital_sign.ITEMID==100]
heartr=surgery_vital_sign[surgery_vital_sign.ITEMID==40]
pulse=surgery_vital_sign[surgery_vital_sign.ITEMID==44]
respirationr=surgery_vital_sign[surgery_vital_sign.ITEMID==92]
spo2=surgery_vital_sign[surgery_vital_sign.ITEMID==188]
vital_sign_data = np.zeros((3386,100,8))
def timeseriesfill(sid,sdate,stime,itemid,n,m):
    global temp
    if itemid==65:
        temp=sbp
    if itemid==66:
        temp=dbp
    if itemid==71:
        temp=cvp
    if itemid==100:
        temp=temperature
    if itemid==40:
        temp=heartr
    if itemid==44:
        temp=pulse
    if itemid==92:
        temp=respirationr
    if itemid==188:
        temp=spo2
    tempp=temp[(temp.SUBJECT_ID==sid)&(temp.MONITOR_DATE==sdate)]
    tempp.reset_index(drop=True,inplace=True)
    end=tempp.shape[0]
    if end==0:
        vital_sign_data[n,:,m]=temp['VALUE'].median()
    if end!=0:
        start=int((tempp.loc[0,'MONITOR_TIME']-stime).total_seconds()/60/5)
        if start<0 and end<=100:
            vital_sign_data[n,0:end+start,m]=tempp.loc[-start:,'VALUE']
            vital_sign_data[n,end+start:,m]=tempp.loc[end-1,'VALUE']
        if start<0 and end>100 and end+start>100:
            vital_sign_data[n,:,m]=tempp.loc[-start:99-start,'VALUE']
        if start<0 and end>100 and end+start<100:
            vital_sign_data[n,:end+start,m]=tempp.loc[-start:,'VALUE']
            vital_sign_data[n,end+start:,m]=tempp.loc[end-1,'VALUE']
        if start==0 and end<=100:
            vital_sign_data[n,0:end,m]=tempp.loc[:,'VALUE']
            vital_sign_data[n,end:,m]=tempp.loc[end-1,'VALUE']
        if start==0 and end>100:
            vital_sign_data[n,:,m]=tempp.loc[:99,'VALUE']
        if start>0 and start+end<=100:
            vital_sign_data[n,0:start,m]=tempp.loc[0,'VALUE']
            vital_sign_data[n,start:start+end,m]=tempp.loc[:,'VALUE']
            vital_sign_data[n,start+end:,m]=tempp.loc[end-1,'VALUE']
        if start>0 and start+end>100:
            vital_sign_data[n,0:start,m]=tempp.loc[0,'VALUE']
            vital_sign_data[n,start:,m]=tempp.loc[:99-start,'VALUE']
def processvitalsign(sid,sdate,stime,n):
    timeseriesfill(sid,sdate,stime,65,n,0)
    timeseriesfill(sid,sdate,stime,66,n,1)
    timeseriesfill(sid,sdate,stime,71,n,2)
    timeseriesfill(sid,sdate,stime,100,n,3)
    timeseriesfill(sid,sdate,stime,40,n,4)
    timeseriesfill(sid,sdate,stime,44,n,5)
    timeseriesfill(sid,sdate,stime,92,n,6)
    timeseriesfill(sid,sdate,stime,188,n,7)
list(map(lambda sid,sdate,stime,n:processvitalsign(sid,sdate,stime,n),
         CHDdata_AKI['INHOSP_INDEX_NO'],CHDdata_AKI['手术日期'],CHDdata_AKI['surgery_start_time'],CHDdata_AKI['ID']))
```
